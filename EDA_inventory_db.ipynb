```python
# Creating summary table including the potential data for vendor_invoice 
freight_summary = pd.read_sql_query("""SELECT 
    VendorNumber, 
    SUM(Freight) AS FreightCost 
FROM vendor_invoice 
GROUP BY VendorNumber""", conn)
freight_summary
```

```python
# Creating summary table including the potential data for purchase table 
purchase_summary = pd.read_sql_query("""SELECT 
    p1.VendorNumber,
    p1.VendorName,
    p1.Brand,
    p1.Description,
    p1.PurchasePrice,
    p2.Price AS ActualPrice,
    p2.Volume,
    SUM(p1.Quantity) AS TotalPurchaseQuantity,
    SUM(p1.Dollars) AS TotalPurchaseDollars
FROM purchases p1
JOIN purchase_prices p2
    ON p1.Brand = p2.Brand
WHERE p1.PurchasePrice > 0
GROUP BY p1.VendorNumber,p1.VendorName, p1.Brand, p1.Description, p1.PurchasePrice, p2.Price, p2.Volume
ORDER BY TotalPurchaseDollars
""",conn)
purchase_summary
```

```python
# Creating a summary table including the potential data from sales table 
sales_summary = pd.read_sql_query("""SELECT 
    VendorNo,
    Brand,
    SUM(SalesQuantity) AS TotalSalesQuantity,
    SUM(SalesDollars) AS TotalSalesDollars,
    SUM(SalesPrice) AS TotalSalesPrice,
    SUM(ExciseTax) AS TotalExciseTax
FROM sales
GROUP BY VendorNo, Brand
ORDER BY TotalSalesDollars""",conn)
sales_summary
```

```python
# Creating an aggregated summary table by merging all the summary tables; freight_summary, purchase_summary, sales_summary

vendor_sales_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT 
        VendorNumber, 
        SUM(Freight) AS FreightCost 
    FROM vendor_invoice 
    GROUP BY VendorNumber
), 

PurchaseSummary AS (
    SELECT 
        p1.VendorNumber,
        p1.VendorName,
        p1.Brand,
        p1.Description,
        p1.PurchasePrice,
        p2.Price AS ActualPrice,
        p2.Volume,
        SUM(p1.Quantity) AS TotalPurchaseQuantity,
        SUM(p1.Dollars) AS TotalPurchaseDollars
    FROM purchases p1
    JOIN purchase_prices p2
        ON p1.Brand = p2.Brand
    WHERE p1.PurchasePrice > 0
    GROUP BY p1.VendorNumber, p1.VendorName, p1.Brand, p1.Description, p1.PurchasePrice, p2.Price, p2.Volume
), 

SalesSummary AS (
    SELECT 
        VendorNo,
        Brand,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
) 

SELECT 
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss 
    ON ps.VendorNumber = ss.VendorNo 
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs 
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""",conn)
```

```python
# changing datatype of volume from object to float
vendor_sales_summary['Volume'] = vendor_sales_summary['Volume'].astype('float64')
```

```python
# filling missing values with 0
vendor_sales_summary.fillna(0,inplace = True)
```

```python
# removing unnecessary spaces from categorical columns
vendor_sales_summary['VendorName'] = vendor_sales_summary['VendorName'].str.strip()
vendor_sales_summary['Description'] = vendor_sales_summary['Description'].str.strip()
```

```python
# creating a new column to showcase GrossProfit 
vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
```

```python
# creating a new column to showcase ProfitMargin
vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars'])*100
```

```python
# creating a new column to showcase StockTurnover
vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity'] / vendor_sales_summary['TotalPurchaseQuantity']
```

```python
# creating a new column to showcase SalesToPurchaseRatio
vendor_sales_summary['SalesToPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars'] / vendor_sales_summary['TotalPurchaseDollars']
```

```python
raw_conn = engine.raw_connection()
cursor = raw_conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS vendor_sales_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(150),
    PurchasePrice DECIMAL(10,2),
    ActualPrice DECIMAL(10,2),
    Volume INT,
    TotalPurchaseQuantity INT,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity INT,
    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    FreightCost DECIMAL(15,2),
    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalesToPurchaseRatio DECIMAL(15,2),
    PRIMARY KEY (VendorNumber, Brand)
);
""")

raw_conn.commit()
cursor.close()
raw_conn.close()
```
